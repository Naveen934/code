```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, TensorDataset
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from collections import Counter

# ==================== SETUP ====================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device}")

# ==================== LOAD DATA ====================
train = pd.read_csv("/content/train.txt", sep=";", names=["text", "sentiment"])
test = pd.read_csv("/content/test.txt", sep=";", names=["text", "sentiment"])
val = pd.read_csv("/content/val.txt", sep=";", names=["text", "sentiment"])

print(f"Train: {len(train)}, Test: {len(test)}, Val: {len(val)}")

# ==================== CREATE VOCABULARY ====================
# Simple text cleaning
def clean_text(text):
    text = str(text).lower()
    text = ''.join([c for c in text if c.isalpha() or c.isspace()])
    return text

train['clean_text'] = train['text'].apply(clean_text)

# Create vocabulary
vectorizer = CountVectorizer(max_features=3000)
vectorizer.fit(train['clean_text'])
vocab_size = len(vectorizer.vocabulary_)
print(f"Vocabulary size: {vocab_size}")

# ==================== PREPARE DATA ====================
# Vectorize text
X_train = vectorizer.transform(train['clean_text']).toarray()
X_test = vectorizer.transform(test['text'].apply(clean_text)).toarray()
X_val = vectorizer.transform(val['text'].apply(clean_text)).toarray()

# Encode labels
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train['sentiment'])
y_test = label_encoder.transform(test['sentiment'])
y_val = label_encoder.transform(val['sentiment'])

num_classes = len(label_encoder.classes_)
print(f"Classes: {label_encoder.classes_}")
print(f"Number of classes: {num_classes}")

# ==================== SIMPLE LSTM MODEL ====================
class SimpleLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super().__init__()

        # Project bag-of-words to LSTM input dimension
        self.embed = nn.Linear(input_size, hidden_size)

        # Single LSTM layer
        self.lstm = nn.LSTM(
            input_size=hidden_size,
            hidden_size=hidden_size,
            batch_first=True,
            dropout=0.3
        )

        # Classifier
        self.fc = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        # x shape: (batch_size, vocab_size)

        # Reshape for LSTM: add sequence dimension
        x = x.unsqueeze(1)  # (batch_size, 1, vocab_size)

        # Project to hidden dimension
        x = self.embed(x)   # (batch_size, 1, hidden_size)

        # LSTM
        lstm_out, (hidden, cell) = self.lstm(x)

        # Get last hidden state
        last_hidden = hidden[-1]  # (batch_size, hidden_size)

        # Classify
        output = self.fc(last_hidden)  # (batch_size, num_classes)

        return output

# Initialize model
model = SimpleLSTM(input_size=vocab_size,hidden_size=128,num_classes=num_classes).to(device)

print("\nModel Summary:")
print(model)
print(f"Total parameters: {sum(p.numel() for p in model.parameters()):,}")

# ==================== DATALOADERS ====================
# Convert to tensors
X_train_tensor = torch.tensor(X_train).float()
y_train_tensor = torch.tensor(y_train)
X_test_tensor = torch.tensor(X_test).float()
y_test_tensor = torch.tensor(y_test)
X_val_tensor = torch.tensor(X_val).float()
y_val_tensor = torch.tensor(y_val)

# Create datasets
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)

# Create dataloaders
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# ==================== TRAINING SETUP ====================
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# ==================== TRAINING FUNCTIONS ====================
def train_one_epoch():
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for batch_x, batch_y in train_loader:
        batch_x, batch_y = batch_x.to(device), batch_y.to(device)

        optimizer.zero_grad()
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        #print("out",outputs.data)
        _, predicted = torch.max(outputs.data, 1)
        #print("pred",predicted)
        total += batch_y.size(0)
        correct += (predicted == batch_y).sum().item()

    return total_loss / len(train_loader), 100 * correct / total

def evaluate(loader):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0

    with torch.no_grad():
        for batch_x, batch_y in loader:
            batch_x, batch_y = batch_x.to(device), batch_y.to(device)

            outputs = model(batch_x)
            loss = criterion(outputs, batch_y)

            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += batch_y.size(0)
            correct += (predicted == batch_y).sum().item()

    return total_loss / len(loader), 100 * correct / total

# ==================== TRAINING LOOP ====================
print("\n" + "="*50)
print("Training Started")
print("="*50)

epochs = 10
best_val_acc = 0

for epoch in range(epochs):
    # Train
    train_loss, train_acc = train_one_epoch()

    # Validate
    val_loss, val_acc = evaluate(val_loader)

    print(f"Epoch {epoch+1}/{epochs}:")
    print(f"  Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%")
    print(f"  Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")

    # Save best model
    if train_acc > best_val_acc:
        best_val_acc = train_acc
        torch.save(model.state_dict(), 'best_simple_lstm.pth')
        print(f"  ↳ Saved best model!")

# ==================== TEST FINAL MODEL ====================
print("\n" + "="*50)
print("Testing Best Model")
print("="*50)

# Load best model
model.load_state_dict(torch.load('best_simple_lstm.pth'))

# Test
test_loss, test_acc = evaluate(test_loader)
print(f"Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.2f}%")

# ==================== PREDICTION FUNCTION ====================
def predict(text):
    model.eval()

    # Clean and vectorize
    clean = clean_text(text)
    vectorized = vectorizer.transform([clean]).toarray()

    # Convert to tensor
    input_tensor = torch.tensor(vectorized).float().to(device)

    # Predict
    with torch.no_grad():
        output = model(input_tensor)
        probabilities = F.softmax(output, dim=1)
        _, predicted = torch.max(output, 1)

    # Get results
    print("label",label_encoder.inverse_transform(predicted.cpu().numpy()))
    sentiment = label_encoder.inverse_transform(predicted.cpu().numpy())[0]
    print("sentiments",sentiment)
    confidence = probabilities[0][predicted].item()

    return sentiment, confidence

# ==================== TEST EXAMPLES ====================
print("\n" + "="*50)
print("Example Predictions")
print("="*50)

test_texts = [
    "I love this movie!",
    "This is terrible.",
    "It was okay, not great.",
    "Absolutely fantastic experience!",
    "Very disappointed with the service."
]

for text in test_texts:
    sentiment, confidence = predict(text)
    print(f'\"\"\"\"{text[:30]}...\"\"\"\" \u2192 {sentiment} ({confidence:.1%})')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from nltk.stem import PorterStemmer
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset,TensorDataset,DataLoader
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

sw=stopwords.words('english')

train=pd.read_csv("/content/train.txt",sep=";",names=["text","label"])
test=pd.read_csv("/content/test.txt",sep=";",names=["text","label"])
val=pd.read_csv("/content/val.txt",sep=";",names=["text","label"])

stem=PorterStemmer()
def clean(text):
   if not isinstance(text,str):
      text=str(text)
   text=text.lower()
   text=' '.join([stem.stem(word) for word in text.split() if word not in set(sw) and not word.isdigit()])
   return text

train['t']=train["text"].apply(clean)
test['t']=test["text"].apply(clean)
val['t']=val["text"].apply(clean)

cv=CountVectorizer(max_features=3000)

x_tr=torch.tensor(cv.fit_transform(train['t']).toarray()).float()
x_te=torch.tensor(cv.transform(test['t']).toarray()).float()
x_val=torch.tensor(cv.transform(val['t']).toarray()).float()

le=LabelEncoder()
y_tr=torch.tensor(le.fit_transform(train['label']))
y_te=torch.tensor(le.transform(test['label']))
y_val=torch.tensor(le.transform(val['label']))


training=TensorDataset(x_tr,y_tr)
testing=TensorDataset(x_te,y_te)
valing=TensorDataset(x_val,y_val)


training=DataLoader(training,batch_size=32,shuffle=True)
testing=DataLoader(testing,batch_size=32,shuffle=True)
valing=DataLoader(valing,batch_size=32,shuffle=True)


class Lstm(nn.Module):
   def __init__(self,input_size,hs,labelsize):
      super().__init__() # Call the parent constructor
      self.emp=nn.Linear(input_size,hs)
      self.lstm=nn.LSTM(hs,hs,batch_first=True,dropout=0.3)
      self.fc=nn.Linear(hs,labelsize)
   def forward(self,x):
      x=x.unsqueeze(1) # Corrected typo: unsqeeze -> unsqueeze
      x=self.emp(x)
      _,(hidden,_)=self.lstm(x)
      output=self.fc(hidden[-1])
      return output

input_size=len(x_tr[0])
hidden_size=128
labelsize=train['label'].nunique()
model=Lstm(input_size,hidden_size,labelsize)

def one_train(training):
   model.train()
   total_loss=0
   correct_predictions = 0
   total_samples = 0

   lossfn=nn.CrossEntropyLoss()
   optimizer=optim.Adam(model.parameters(),lr=0.01)
   for x_batch,y_batch in training: # Adjusted for DataLoader yielding tuples

       optimizer.zero_grad()
       outputs=model(x_batch)

       loss=lossfn(outputs,y_batch) # Use raw outputs for CrossEntropyLoss
       loss.backward()
       optimizer.step()

       total_loss+=loss.item()

       _, predicted = torch.max(outputs.data, 1)
       total_samples += y_batch.size(0)
       correct_predictions += (predicted == y_batch).sum().item()

   avg_loss = total_loss / total_samples # This is average loss, not accuracy
   acc = 100 * correct_predictions / total_samples if total_samples > 0 else 0 # Corrected accuracy calculation

   return avg_loss,acc

def evalmodel(testing): # Changed x,y to x_loader, y_loader as they are DataLoaders
   model.eval()
   total_loss = 0
   total_sample = 0
   correct_predictions = 0
   lossfn = nn.CrossEntropyLoss()

   with torch.no_grad():
     for x_batch, y_batch in testing:

        outputs = model(x_batch)
        _,predicted=torch.max(outputs.data,1)
        loss = lossfn(outputs, y_batch)
        total_loss += loss.item()
        total_sample+=y_batch.size(0)
        correct_predictions += (predicted==y_batch).sum().item()

   return total_loss / total_sample , 100*correct_predictions/total_sample

epocs=10
best_acc=0
for epoc in range(epocs):
    avg_loss,acc=one_train(training) # Renamed accuracy to avg_loss to avoid confusion
    if best_acc<acc:
        best_acc=acc
        torch.save(model.state_dict(),'best_model.pth')
        print("best model saved")
    avg_loss_v,acc_v=evalmodel(testing)

    print(f"epoch{epoc+1}/{epocs}:")
    print(f"Train accuracy : {acc:.2f}%", f"Train avg_loss: {avg_loss:.4f}", f"val accuracy: {acc_v:.2f}", f"Val loss : {avg_loss_v:.4f}")


```
