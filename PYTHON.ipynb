```python
"""Operation	Primary Use Case	Joins On	Best For
concat	Stacking	Index/Position	Combining similar structured data
merge	Database joins	Columns	SQL-style relationships
join	Index-based joins	Index	Time series, indexed data

Use concat when you want to stack DataFrames with same columns (append rows) or same rows (append columns)
Use merge when you have common columns and want database-style joins (most common for relational data)
Use join when you want to combine DataFrames based on their indexes"""

"""1. pd.concat() - Stacking Data
Purpose: Stack DataFrames vertically (rows) or horizontally (columns)"""

import pandas as pd
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']})
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'B': ['B2', 'B3']})
# Vertical stacking (axis=0)
result = pd.concat([df1, df2], axis=0)
#     A   B
# 0  A0  B0
# 1  A1  B1
# 0  A2  B2
# 1  A3  B3
# Horizontal stacking (axis=1)
result = pd.concat([df1, df2], axis=1)
#     A   B   A   B
# 0  A0  B0  A2  B2
# 1  A1  B1  A3  B3


"""2. pd.merge() - Database-style Joins
Purpose: Combine DataFrames based on common columns (like SQL JOIN)"""

left = pd.DataFrame({'N': ['K0', 'K1', 'K2'], 'A': ['A0', 'A1', 'A2']})
right = pd.DataFrame({'N': ['K0', 'K1', 'K3'], 'B': ['B0', 'B1', 'B3']})
# Inner join (default)
result = pd.merge(left, right, on='N')
#   key   A   B
# 0  K0  A0  B0
# 1  K1  A1  B1
# Left join
result = pd.merge(left, right, on='N', how='left')
#   key   A    B
# 0  K0  A0   B0
# 1  K1  A1   B1
# 2  K2  A2  NaN
# Right join
result = pd.merge(left, right, on='N', how='right')
#   key    A   B
# 0  K0   A0  B0
# 1  K1   A1  B1
# 2  K3  NaN  B3
# Outer join
result = pd.merge(left, right, on='N', how='outer')
#   key    A    B
# 0  K0   A0   B0
# 1  K1   A1   B1
# 2  K2   A2  NaN
# 3  K3  NaN   B3


"""3. df.join() - Index-based Joins
Purpose: Join DataFrames based on their indexes"""

left = pd.DataFrame({'A': ['A0', 'A1', 'A2']}, index=['I0', 'I1', 'I2'])
right = pd.DataFrame({'B': ['B0', 'B1', 'B2']}, index=['I0', 'I2', 'I3'])

# Default left join on index
result = left.join(right)
#      A    B
# I0  A0   B0
# I1  A1  NaN
# I2  A2   B2

# Right join
result = left.join(right, how='right')
#       A   B
# I0   A0  B0
# I2   A2  B2
# I3  NaN  B2

# Inner join
result = left.join(right, how='inner')
#     A   B
# I0 A0  B0
# I2 A2  B2

pd.concat([df1, df2], axis=1)  # ≈ df1.join(df2) when indexes align
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def eval_regression_model(y_true, y_pred, model_name="Model"):
    """
    Comprehensive regression model evaluation
    
    Parameters:
    - y_true: Actual target values
    - y_pred: Predicted target values
    - model_name: Name of the model for display purposes
    
    Returns:
    - Dictionary containing all metrics
    """
    
    # Calculate metrics
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    # Additional metrics
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100  # Mean Absolute Percentage Error
    adjusted_r2 = 1 - (1 - r2) * (len(y_true) - 1) / (len(y_true) - len(y_true) - 1)
    
    # Print results
    print(f"╔{'═' * 50}╗")
    print(f"║           REGRESSION MODEL EVALUATION           ║")
    print(f"║                 {model_name:^20}                 ║")
    print(f"╠{'═' * 50}╣")
    print(f"║ MAE  : {mae:10.4f} (Mean Absolute Error)        ║")
    print(f"║ MSE  : {mse:10.4f} (Mean Squared Error)         ║")
    print(f"║ RMSE : {rmse:10.4f} (Root Mean Squared Error)   ║")
    print(f"║ R²   : {r2:10.4f} (R-squared)                   ║")
    print(f"║ MAPE : {mape:9.2f}% (Mean Absolute Percentage)  ║")
    print(f"║ Adj R²: {adjusted_r2:8.4f} (Adjusted R-squared)     ║")
    print(f"╚{'═' * 50}╝")
    
    # Return metrics dictionary
    metrics = {
        'MAE': mae,
        'MSE': mse, 
        'RMSE': rmse,
        'R2': r2,
        'MAPE': mape,
        'Adjusted_R2': adjusted_r2,
        'Model_Name': model_name
    }
    
    return metrics

def plot_regression_evaluation(y_true, y_pred, model_name="Model"):
    """
    Visualize regression model performance
    """
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot 1: Actual vs Predicted
    axes[0].scatter(y_true, y_pred, alpha=0.6)
    axes[0].plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', lw=2)
    axes[0].set_xlabel('Actual Values')
    axes[0].set_ylabel('Predicted Values')
    axes[0].set_title(f'{model_name} - Actual vs Predicted')
    axes[0].grid(True, alpha=0.3)
    
    # Plot 2: Residuals
    residuals = y_true - y_pred
    axes[1].scatter(y_pred, residuals, alpha=0.6)
    axes[1].axhline(y=0, color='r', linestyle='--')
    axes[1].set_xlabel('Predicted Values')
    axes[1].set_ylabel('Residuals')
    axes[1].set_title(f'{model_name} - Residual Plot')
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return residuals

# Test regression evaluation
if __name__ == "__main__":
    # Sample data
    np.random.seed(42)
    y_true_reg = np.random.normal(50, 15, 100)
    y_pred_reg = y_true_reg + np.random.normal(0, 5, 100)
    
    metrics = eval_regression_model(y_true_reg, y_pred_reg, "Linear Regression")
    residuals = plot_regression_evaluation(y_true_reg, y_pred_reg, "Linear Regression")

    
```

```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                           f1_score, roc_auc_score, confusion_matrix, 
                           classification_report, roc_curve, precision_recall_curve)
import matplotlib.pyplot as plt
import seaborn as sns

def eval_classification_model(y_true, y_pred, y_pred_proba=None, model_name="Model"):
    """
    Comprehensive classification model evaluation
    
    Parameters:
    - y_true: Actual labels
    - y_pred: Predicted labels  
    - y_pred_proba: Predicted probabilities (for ROC/AUC)
    - model_name: Name of the model for display purposes
    
    Returns:
    - Dictionary containing all metrics
    """
    
    # Basic metrics
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted')
    recall = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')
    
    # Confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    
    # ROC-AUC (if probabilities are provided)
    roc_auc = None
    if y_pred_proba is not None:
        if len(np.unique(y_true)) == 2:  # Binary classification
            roc_auc = roc_auc_score(y_true, y_pred_proba)
        else:  # Multi-class
            roc_auc = roc_auc_score(y_true, y_pred_proba, multi_class='ovr')
    
    # Print results
    print(f"╔{'═' * 60}╗")
    print(f"║              CLASSIFICATION MODEL EVALUATION             ║")
    print(f"║                   {model_name:^20}                   ║")
    print(f"╠{'═' * 60}╣")
    print(f"║ Accuracy  : {accuracy:8.4f}                            ║")
    print(f"║ Precision : {precision:8.4f} (Weighted)                ║")
    print(f"║ Recall    : {recall:8.4f} (Weighted)                  ║")
    print(f"║ F1-Score  : {f1:8.4f} (Weighted)                    ║")
    if roc_auc is not None:
        print(f"║ ROC-AUC   : {roc_auc:8.4f}                            ║")
    print(f"╚{'═' * 60}╝")
    
    # Detailed classification report
    print("\n" + "="*50)
    print("DETAILED CLASSIFICATION REPORT")
    print("="*50)
    print(classification_report(y_true, y_pred))
    
    # Return metrics dictionary
    metrics = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1_Score': f1,
        'ROC_AUC': roc_auc,
        'Confusion_Matrix': cm,
        'Model_Name': model_name
    }
    
    return metrics

def plot_classification_evaluation(y_true, y_pred, y_pred_proba=None, model_name="Model"):
    """
    Visualize classification model performance
    """
    n_plots = 2 if y_pred_proba is None else 3
    fig, axes = plt.subplots(1, n_plots, figsize=(5*n_plots, 4))
    
    if n_plots == 2:
        axes = [axes] if n_plots == 1 else axes
    
    # Plot 1: Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0])
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('Actual')
    axes[0].set_title(f'{model_name} - Confusion Matrix')
    
    # Plot 2: Classification Metrics Comparison
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
    scores = [
        accuracy_score(y_true, y_pred),
        precision_score(y_true, y_pred, average='weighted'),
        recall_score(y_true, y_pred, average='weighted'),
        f1_score(y_true, y_pred, average='weighted')
    ]
    
    axes[1].bar(metrics, scores, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
    axes[1].set_ylim(0, 1)
    axes[1].set_title(f'{model_name} - Metrics Comparison')
    axes[1].set_ylabel('Score')
    
    # Plot 3: ROC Curve (if probabilities available)
    if y_pred_proba is not None and len(np.unique(y_true)) == 2:
        fpr, tpr, _ = roc_curve(y_true, y_pred_proba)
        roc_auc = roc_auc_score(y_true, y_pred_proba)
        
        axes[2].plot(fpr, tpr, color='darkorange', lw=2, 
                    label=f'ROC curve (AUC = {roc_auc:.2f})')
        axes[2].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        axes[2].set_xlim([0.0, 1.0])
        axes[2].set_ylim([0.0, 1.05])
        axes[2].set_xlabel('False Positive Rate')
        axes[2].set_ylabel('True Positive Rate')
        axes[2].set_title(f'{model_name} - ROC Curve')
        axes[2].legend(loc="lower right")
        axes[2].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Test classification evaluation
if __name__ == "__main__":
    # Sample binary classification data
    np.random.seed(42)
    y_true_clf = np.random.randint(0, 2, 100)
    y_pred_clf = np.random.randint(0, 2, 100)
    y_pred_proba = np.random.rand(100)
    
    metrics = eval_classification_model(y_true_clf, y_pred_clf, y_pred_proba, "Random Forest")
    plot_classification_evaluation(y_true_clf, y_pred_clf, y_pred_proba, "Random Forest")
```

```python
class ModelEvaluator:
    """
    Comprehensive model evaluation for both regression and classification
    """
    
    def __init__(self):
        self.regression_metrics = {}
        self.classification_metrics = {}
    
    def evaluate_regression(self, y_true, y_pred, model_name="Regression Model"):
        """Evaluate regression model with comprehensive metrics"""
        return eval_regression_model(y_true, y_pred, model_name)
    
    def evaluate_classification(self, y_true, y_pred, y_pred_proba=None, model_name="Classification Model"):
        """Evaluate classification model with comprehensive metrics"""
        return eval_classification_model(y_true, y_pred, y_pred_proba, model_name)
    
    def compare_models(self, models_metrics, model_type='regression'):
        """
        Compare multiple models
        models_metrics: List of metrics dictionaries from evaluate_* methods
        """
        if model_type == 'regression':
            comparison_df = pd.DataFrame(models_metrics)
            print("╔" + "═"*70 + "╗")
            print("║" + "MODEL COMPARISON (REGRESSION)".center(70) + "║")
            print("╠" + "═"*70 + "╣")
        else:
            comparison_df = pd.DataFrame(models_metrics)
            print("╔" + "═"*70 + "╗")
            print("║" + "MODEL COMPARISON (CLASSIFICATION)".center(70) + "║")
            print("╠" + "═"*70 + "╣")
        
        print(comparison_df.to_string(index=False))
        print("╚" + "═"*70 + "╝")
        
        return comparison_df

# Example usage with multiple models
def demo_complete_evaluation():
    """Demonstrate complete evaluation workflow"""
    evaluator = ModelEvaluator()
    
    # Regression example
    print("REGRESSION EVALUATION DEMO")
    print("="*50)
    
    np.random.seed(42)
    y_true_reg = np.random.normal(50, 15, 100)
    
    # Simulate different model predictions
    y_pred_model1 = y_true_reg + np.random.normal(0, 3, 100)  # Good model
    y_pred_model2 = y_true_reg + np.random.normal(0, 10, 100) # Poor model
    
    metrics1 = evaluator.evaluate_regression(y_true_reg, y_pred_model1, "Model 1 (Good)")
    metrics2 = evaluator.evaluate_regression(y_true_reg, y_pred_model2, "Model 2 (Poor)")
    
    # Compare models
    evaluator.compare_models([metrics1, metrics2], 'regression')
    
    # Classification example
    print("\n\nCLASSIFICATION EVALUATION DEMO")
    print("="*50)
    
    y_true_clf = np.random.randint(0, 2, 100)
    y_pred_model1_clf = y_true_clf  # Perfect predictions
    y_pred_model2_clf = np.random.randint(0, 2, 100)  # Random predictions
    y_pred_proba = np.random.rand(100)
    
    metrics1_clf = evaluator.evaluate_classification(y_true_clf, y_pred_model1_clf, 
                                                   y_pred_proba, "Perfect Model")
    metrics2_clf = evaluator.evaluate_classification(y_true_clf, y_pred_model2_clf, 
                                                   y_pred_proba, "Random Model")
    
    # Compare models
    evaluator.compare_models([metrics1_clf, metrics2_clf], 'classification')

# Run demonstration
if __name__ == "__main__":
    demo_complete_evaluation()
```

###Important Dunder Methods in Python (Magic Methods)

1️⃣ Container & Sequence Methods
Method	Purpose	Example Usage
__getitem__(self, key)	Enables obj[key]	obj[0]
__setitem__(self, key, value)	Enables obj[key] = value	obj[0] = 100
__delitem__(self, key)	Enables del obj[key]	del obj[2]
__contains__(self, item)	Enables item in obj	"a" in obj
__len__(self)	Enables len(obj)	len(obj)


```python
class MyList:
    def __init__(self, data):
        self.data = data

    def __getitem__(self, index):
        return self.data[index]  # Enables obj[index]

    def __len__(self):
        return len(self.data)  # Enables len(obj)

obj = MyList([10, 20, 30, 40])
print(obj[2])
print(len(obj))
```

2️⃣ Iteration & Looping
Method	Purpose	Example Usage
__iter__(self)	Enables iteration (for x in obj)	for x in obj:
__next__(self)	Defines behavior for next(obj)	next(obj)

```python
class Counter:
    def __init__(self, max):
        self.max = max
        self.count = 0

    def __iter__(self):
        return self  # Returns self as an iterator

    def __next__(self):
        if self.count >= self.max:
            raise StopIteration  # Stops iteration
        self.count += 1
        return self.count

counter = Counter(3)
for num in counter:
    print(num)  # Output: 1 2 3
```

3️⃣ Arithmetic & Operator Overloading
Method	Purpose	Example Usage
__add__(self, other)	obj1 + obj2	a + b
__sub__(self, other)	obj1 - obj2	a - b
__mul__(self, other)	obj1 * obj2	a * b
__truediv__(self, other)	obj1 / obj2	a / b
__floordiv__(self, other)	obj1 // obj2	a // b
__mod__(self, other)	obj1 % obj2	a % b

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)  # Custom addition

a = Number(5)
b = Number(10)
c = a + b  # Calls __add__
print(c.value)  # Output: 15
```

4️⃣ Comparisons & Sorting
Method	Purpose	Example Usage
__eq__(self, other)	Enables obj1 == obj2	a == b
__ne__(self, other)	Enables obj1 != obj2	a != b
__lt__(self, other)	Enables obj1 < obj2	a < b
__le__(self, other)	Enables obj1 <= obj2	a <= b
__gt__(self, other)	Enables obj1 > obj2	a > b
__ge__(self, other)	Enables obj1 >= obj2	a >= b

```python
class Person:
    def __init__(self, age):
        self.age = age

    def __lt__(self, other):
        return self.age < other.age  # Defines < behavior

p1 = Person(25)
p2 = Person(30)
print(p1 < p2)  # Output: True
```

5️⃣ Object Representation & Debugging
Method	Purpose	Example Usage
__str__(self)	Defines string representation for print(obj)	print(obj)
__repr__(self)	Defines official string representation (for debugging)	repr(obj)

```python
class Car:
    def __init__(self, model):
        self.model = model

    def __str__(self):
        return f"Car Model: {self.model}"  # For print()

    def __repr__(self):
        return f"CarNa(veen'{self.model}')"  # For debugging

car = Car("Tesla")
print(car)  # Output: Car Model: Tesla
print(repr(car))  # Output: Car('Tesla')
```

__call__(self, *args, **kwargs)	Allows object to be called like a function	obj()


```python
class Adder:
    def __call__(self, a, b):
        return a + b  # Makes the object callable

add = Adder()
print(add(5, 3))  # Output: 8
```

7️⃣ Object Creation & Memory Management
Method	Purpose	Example Usage
__new__(cls, *args, **kwargs)	Controls object creation	obj = cls()
__init__(self, *args, **kwargs)	Initializes the object	obj = cls()
__del__(self)	Called when an object is deleted	del obj

```python
class Singleton:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance

s1 = Singleton()
s2 = Singleton()
print(s1 is s2)  # Output: True (same instance)
```
